```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
viz=pd.read_csv('output/agg_data.csv')
```

```python
dealership1=viz[viz['dealership_id']=='dealership1']
dealership2=viz[viz['dealership_id']=='dealership2']
```

```python
fig, ax = plt.subplots(figsize=[16, 9])
ax.plot(dealership1['months'].sort_values(),dealership1['car_nums'],label='dealership1')
ax.plot(dealership2['months'].sort_values(),dealership2['car_nums'],label='dealership2')
ax.set_xlabel('months')
ax.set_ylabel('# of cars')
plt.title('Car Inventory Per Dealership 2021')
plt.legend()
plt.show()
plt.savefig('output/viz.png')
```
